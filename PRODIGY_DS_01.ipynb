### Create a bar chart or histogram to visualize the distribution of  a categorical or continuous variable , such as the distribution of ages or genders in a population

```python
import numpy as np
import pandas as pd
data=pd.read_csv("task1.csv",skiprows=4)
data.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
population_2020 = data['2020'].dropna()
# Create a histogram to visualize the distribution of the population in 2020
plt.figure(figsize=(10, 6))
plt.hist(population_2020,color='red', bins=30, edgecolor='black')
plt.title('Distribution of Population by Country in 2020')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.yscale('log') # Use a logarithmic scale for better visibility
plt.grid(True)
plt.show()
```

```python
#the bar chart showing the population of Afghanistan from 1960 to 2022:
import pandas as pd
import matplotlib.pyplot as plt
# Select data for Afghanistan
afghanistan_data = data[data['Country Name'] == 'Afghanistan'].iloc[0, 4:-2]
# Convert the series to numeric values and drop NaNs
afghanistan_data = pd.to_numeric(afghanistan_data, errors='coerce').dropna()
# Plotting the population data over the years
plt.figure(figsize=(14, 7))
plt.bar(afghanistan_data.index, afghanistan_data.values, color='green')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population of Afghanistan (1960-2022)')
plt.xticks(rotation=45)
plt.tight_layout()
# Display the plot
plt.show()
```

```python
#. Population Comparison of Multiple Countries in 2020
# List of countries to compare
countries = ['Afghanistan', 'India', 'United States', 'China', 'Brazil']
# Select population data for the year 2020
population_2020 = data[data['Country Name'].isin(countries)][['Country Name', '2020']]
# Plotting the population data
plt.figure(figsize=(12, 6))
plt.bar(population_2020['Country Name'], population_2020['2020'], color='')
plt.xlabel('Country')
plt.ylabel('Population in 2020')
plt.title('Population Comparison of Selected Countries in 2020')
plt.xticks(rotation=45)
plt.tight_layout()
# Display the plot
plt.show()
```

```python
# Population Growth Over Time for Multiple Countries

# List of countries to compare
countries = ['Afghanistan', 'India', 'United States', 'China', 'Brazil']

# Select population data for the chosen countries
population_data = data[data['Country Name'].isin(countries)].set_index('Country Name')
# Years to include in the plot
years = population_data.columns[4:-2]
# Plotting the population data as histograms over the years
plt.figure(figsize=(14, 7))
for country in countries:
    country_data = pd.to_numeric(population_data.loc[country, years], errors='coerce').dropna()
    plt.hist(country_data.values, bins=20, alpha=0.5, label=country)
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.title('Population Distribution for Selected Countries (1960-2022)')
plt.legend()
plt.tight_layout()
# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt



# Select data for Africa Eastern and Southern
region = 'Africa Eastern and Southern'
region_data = data[data['Country Name'] == region]

# Select population data for years (assuming column names start from year 1960)
population_columns = [col for col in region_data.columns if col.isdigit() and int(col) >= 1960]
region_data = region_data[population_columns]

# Handle missing values (consider both non-numeric and NaNs)
def handle_missing_values(df):
    # Drop columns with all NaN values
    df = df.dropna(axis=1, how='all')
    # Convert to numeric, coercing errors to NaN
    df = df.apply(pd.to_numeric, errors='coerce')
    # Replace remaining NaNs with the mean of the column
    df = df.fillna(df.mean())
    return df

region_data = handle_missing_values(region_data.copy())  # Avoid modifying original data

# Calculate the average population across all years
average_population = region_data.mean(axis=1).mean()

# Plotting the average population
plt.figure(figsize=(14, 7))
plt.bar([region], [average_population], color='skyblue')
plt.xlabel('Region')
plt.ylabel('Average Population (1960-2022)')
plt.title('Average Population of Africa Eastern and Southern (1960-2022)')
plt.tight_layout()

# Display the plot
plt.grid(axis='y')
plt.show()
```

```python
# List of countries to compare
countries = ['Afghanistan', 'India', 'United States', 'China', 'Brazil']
# Select population data for the chosen countries
population_data = data[data['Country Name'].isin(countries)].set_index('Country Name')
# Years to include in the plot
years = population_data.columns[4:-2]
# Plotting the population data as histograms over the years
fig, axes = plt.subplots(len(countries), 1, figsize=(10, 8), sharex=True)
for i, country in enumerate(countries):
    country_data = pd.to_numeric(population_data.loc[country, years], errors='coerce').dropna()
    axes[i].hist(country_data.values, bins=20, alpha=0.7)
    axes[i].set_title(f'Population Distribution for {country} (1960-2022)')
    axes[i].set_xlabel('Population')
    axes[i].set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```
